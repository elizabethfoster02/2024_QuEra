# iQuHack Quera Challenge Submission Documentation
#### Dirac Dawgs (Masarah Ahmedhussain, Elizabeth Foster, Avery Hanna, Noah Stiegler, Temazulu Zulu)

The Challenge: Explore a possible route to quantum advantage through finding maximum independent sets in higher order connectivity square lattice graphs

QuEra's neutral atom quantum computer is uniquely suited to finding the maximum independent set (MIS) of a unit disk (UD) graph. In this case, unit disk graphs are those generated by placing a series of nodes or points on a 2d plane, drawing circles of equal radius around them, and connecting the nodes which are within each other's circles via edges or lines. A maximum independent set is a set containing the maximum possible number of points, where no two points in the set are within each other's disks.
